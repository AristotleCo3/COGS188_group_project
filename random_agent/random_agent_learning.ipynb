```python
import numpy as np
import random
import matplotlib.pyplot as plt
from minesweeper import core
from randomAgent import RandomMinesweeperAgent

# Initialize the Minesweeper Environment
env = core.Board(rows=10, cols=10, mines=30)

# Run the Random Agent
random_agent = RandomMinesweeperAgent(env, num_episodes=1000)
random_agent.play()

# Plot Results
random_agent.plot_rewards()
random_agent.plot_moves()
random_agent.print_summary()
```
